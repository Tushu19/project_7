```python
g=[]  #stores all converted value in ascii
count=0 #0
for j in range (0,len(diff_array),8):
  g.append(chr(sign_lst[count]))
  iter=0
  k0=0
  z0=0
  u0=0
  for i in range (0,8,2):
    #if both are zero do forward grp only
    if((diff_array[j+i])==0 and diff_array[j+i+1]==0):
      #forward grouping
        g.append(chr(diff_array[j+i]*100+diff_array[j+i+1]))
        k0=k0+(2**(3-iter))
        iter+=1
    else:
      #both are not zero check all three grouping styles
      if((diff_array[j+i])!=0 and diff_array[j+i+1]!=0):
        #forward grouping
        if((diff_array[j+i]*100+diff_array[j+i+1])<255):
            g.append(chr(diff_array[j+i]*100+diff_array[j+i+1]))
            k0=k0+(2**(3-iter))
            iter+=1
        #backward grouping
        elif((diff_array[j+i+1])*100+diff_array[j+i]<255):
            g.append(chr(diff_array[j+i+1]*100+diff_array[j+i]))
            z0=z0+(2**(3-iter))
            iter+=1
        #no grouping for both value non zero
        else:
          g.append(chr(diff_array[j+i]))  
          g.append(chr(diff_array[j+i+1]))   
          u0=u0+(2**(3-iter))
          iter+=1
      #no grouping for any one value as zero in pair
      else:
          g.append(chr(diff_array[j+i]))  
          g.append(chr(diff_array[j+i+1]))   
          u0=u0+(2**(3-iter))
          iter+=1
  g.append(chr(k0+100))
  g.append(chr(z0+100))
  g.append(chr(u0+100))
  g.append(chr(r[count]))
  g.append(chr(s[count]))
  g.append(chr(rs[count]))
  g.append(0)
  count=count+1
g.append(chr(ii))
```

```python
#splitting values at zero
diff_array_c=[]
sign_bit_c=[]
rs=[]
r=[]
flag=0
i=0
while(i<len(lines)-8):
    temp=[]
    while(lines[i]!=48 and i<len(lines)-2):
        temp.append(lines[i])
        i=i+1
    i=i+1         #skip 48
    len_temp=len(temp)
    if(len_temp!=0):
        flag=flag+1
        j=len_temp-8
        len_temp=len(temp)
        k=int(bin(temp[len_temp-6]-100).replace("0b", ""))
        z=int(bin(temp[len_temp-5]-100).replace("0b", ""))
        u=int(bin(temp[len_temp-4]-100).replace("0b", ""))
        r.append(int(bin(temp[len_temp-3]+temp[len_temp-2]).replace("0b", "")))
        if(temp[len_temp-1]):
            sign_bit_c.append(int(bin(temp[0]+1).replace("0b", "")))
        else:
            sign_bit_c.append(int(bin(temp[0]).replace("0b", "")))
        stk=[]#stack to store diff array var temporarily
        count=1

        while(count<5):
            if(k%10!=0):
                stk.append(int(temp[j]%100))
                stk.append(int(temp[j]/100))
            elif(z%10!=0):
                stk.append(int(temp[j]/100))
                stk.append(int(temp[j]%100))
            elif(u%10!=0): 
                stk.append(int(temp[j]))
                stk.append(int(temp[j-1]))
                j=j-1
            k=int(k/10)
            u=int(u/10)
            z=int(z/10)  
            j=j-1
            count=count+1
        for d in range (8):
                diff_array_c.append(stk.pop())
ii=lines[len(lines)-1]
print(diff_array_c[0])
diff_array_c[0]=(diff_array_c[0])+(255*ii)
len(diff_array_c)
```
